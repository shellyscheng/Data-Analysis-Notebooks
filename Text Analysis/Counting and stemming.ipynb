# Stop words or stopwords

```python
vec = CountVectorizer(stop_words=['a','the', 'and'])
matrix = vec.fit_transform(texts)

results = pd.DataFrame(matrix.toarray(), columns = vec.get_feature_names())
results

# no column for "the" now
```

```python
vec = CountVectorizer(stop_words='english')
# Stopwords in English (automatically)

matrix = vec.fit_transform(texts)

results = pd.DataFrame(matrix.toarray(), columns = vec.get_feature_names())
results
```

# Steaming

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
```

```python
porter_stemmer.stem('fishing')
```

```python
porter_stemmer.stem('fished')
```

## Stemmer Tokenizer

```python
from textblob import TextBlob

# Use textblob
def textblob_tokenizer(str_input):
    blob = TextBlob(str_input.lower())
    tokens = blob.words
    words = [token.stem() for token in tokens]
    return words

# Use 
def stemming_tokenizer(str_input):
    words = re.sub(r"[^A-Za-z0-9\-]", " ", str_input).lower().split()
    words = [porter_stemmer.stem(word) for word in words]
    return words
```

```python
stemming_tokenizer('I went finshing to get fishes')
```

```python
textblob_tokenizer('I went finshing to get fishes')
```

## Use our tokenizer that also stems

```python
vec = CountVectorizer(stop_words='english', tokenizer=textblob_tokenizer)
# Stopwords in English (automatically)

matrix = vec.fit_transform(texts)

results = pd.DataFrame(matrix.toarray(), columns = vec.get_feature_names())
results
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vec = TfidfVectorizer(stop_words='english', 
                     tokenizer=textblob_tokenizer,
                     use_idf=False,
                     norm='l1')
# Stopwords in English (automatically)

matrix = vec.fit_transform(texts)

results = pd.DataFrame(matrix.toarray(), columns = vec.get_feature_names())
results
```

```python
results.sort_values(by='fish', ascending=False)
```

```python
# Only search for fish and bug now

vec = TfidfVectorizer(stop_words='english', 
                      vocabulary=['fish','bug'],
                      tokenizer=textblob_tokenizer,
                      use_idf=False,
                      norm='l1')
# Stopwords in English (automatically)

matrix = vec.fit_transform(texts)

results = pd.DataFrame(matrix.toarray(), columns = vec.get_feature_names())
results
```

```python
# Only search for fish and bug now

vec = CountVectorizer(stop_words='english', 
                      vocabulary=['fish','bug'],
                      tokenizer=textblob_tokenizer,
                      )
# Stopwords in English (automatically)

matrix = vec.fit_transform(texts)

results = pd.DataFrame(matrix.toarray(), columns = vec.get_feature_names())
results
```
