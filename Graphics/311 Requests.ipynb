### Creating new columns

Create new columns called `created_at` and `closed_at` from the `Created Date` and `Closed Date` columns. Do them individually, and **time them**. datetime

```python
%%time
df['created_at'] = pd.to_datetime(df['Created Date'])
```

```python
%%time
df['created_at'] = pd.to_datetime(df['Created Date'], format='%m/%d/%Y %I:%M:%S %p')
```

```python
%%time
df['closed_at'] = pd.to_datetime(df['Closed Date'], format='%m/%d/%Y %X %p')
```
